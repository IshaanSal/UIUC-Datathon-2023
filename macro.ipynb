```python
rows = 25
print(df['M_FGDP.IUSA'][6])

#df['M_FGDP.IUSA'][9]
#float(df['M_FGDP.IUSA'][9])
#val + 1

df['gdp_diff'] = np.zeros(32)

for i in range(rows):
  if (i+8 < 32):
    val = float(df['M_FGDP.IUSA'][i+8]) - float(df['M_FGDP.IUSA'][i+7])
    val = abs(val)
    perc_diff = val/float(df['M_FGDP.IUSA'][i+7])
    df['gdp_diff'][i+7] = perc_diff
```
